```python

from sklearn.feature_selection import VarianceThreshold

# Remove features with very low variance
selector = VarianceThreshold(threshold=0.01)  # or 0.001 for finer filter
X_var = selector.fit_transform(df)
selected_columns = df.columns[selector.get_support()]
```

```python
selected_columns.shape
```

```python
import seaborn as sns
import numpy as np

corr_matrix = df[selected_columns].corr().abs()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]
X_reduced = df[selected_columns].drop(columns=to_drop)
```

```python
X_train = X_reduced
Y_train = df['Target_Variable/Total Income']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(data=scaler.fit_transform(X_train),columns=X_train.columns)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train_scaled, Y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_tr, y_tr)

importances = model.feature_importances_
feat_importance = pd.Series(importances, index=X_reduced.columns)
top_features = feat_importance.sort_values(ascending=False).head(40).index
```

```python

X_final_tr = X_tr[top_features]
X_final_val = X_val[top_features]  
```

```python
X_final_tr
```

```python
X_final_val
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(X_final_tr,y_tr)

y_final_val_pred = model.predict(X_final_val)
```

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error
print(mean_absolute_percentage_error(y_final_val_pred , y_val))
print(r2_score(y_final_val_pred,y_val))
```
